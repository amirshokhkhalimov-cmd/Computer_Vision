```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4])
y = np.array([10,20,25,30])

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
from matplotlib.lines import lineStyles
x = np.array([1,2,3,4])
y = np.array([4,6,1,15])

plt.plot(y, linestyle='dotted',color='green', marker='o')
plt.title("Styled Line Plot")
plt.show()
```

```python
x = np.array([1,2,3,4])
y1 = np.array([2,4,6,8])
y2 = np.array([1,3,5,7])

plt.plot(x, y1, label = 'line1', color='red',marker='s')
plt.plot(x, y2, label = 'line2', color='green',linestyle='--',marker='o')
plt.title("Multiple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

# Bar chart and numpy **matplotlib** **italicized text**

```python
x = np.array(['A','B', 'C','D'])
y = np.array([3, 7, 10, 8])

plt.bar(x,y, color='yellow')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel(" Values")
plt.show()
```

# Scatter Plot with Numpy and Matplotlib

```python
x = np.array([2,4,6,8])
y = np.array([1,3,5,7])

plt.scatter(x,y, color='purple')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel(" Y")
plt.show()
```

# Histogram with numpy and Matplotlib

```python
data = np.array([3,4,5,6,8,8,8,9,10])

plt.hist(data, color='blue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel(" Count")
plt.show()
```

```python
x= np.array([1,2,3,4])
y1 = np.array([2,4,6,8])
y2 = np.array([3,6,9,12])

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot(x,y1,color='red')
plt.title("Subplot 1")

plt.subplot(1,2,2)
plt.bar(x,y2,color='blue')
plt.title("Subplot 2")

plt.tight_layout()

plt.savefig('plot.png')

plt.show()
```
